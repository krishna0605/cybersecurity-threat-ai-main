```python
from scipy.io import arff
import pandas as pd

# Load ARFF file correctly
data, meta = arff.loadarff("data/KDDTrain+.arff")

# Convert to Pandas DataFrame
df = pd.DataFrame(data)

# Decode byte strings (Fix: Use .map() instead of applymap)
df = df.map(lambda x: x.decode() if isinstance(x, bytes) else x)

# Save as CSV
df.to_csv("data/KDDTrain+.csv", index=False)

print("âœ… Conversion complete! Saved as data/KDDTrain+.csv")

```
